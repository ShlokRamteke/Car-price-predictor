Changing the age from numberic value to a range

```python
def age_groups_creation(row):
    if row['age'] < 30:
        return '20-30'
    elif row['age'] < 40:
        return '30-40'
    elif row['age'] < 50:
        return '40-50'
    elif row['age'] < 60:
        return '50-60'
    else:
        return "60-70"
```

```python
data['age range'] = data.apply( age_groups_creation, axis=1)
```

## Starting Feature Engineering

```python
data.columns
```

```python
X=data[['age', 'annual Salary', 'net worth']]
Y=data[['car purchase amount']]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler    
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
def modelresults(prediction):
    print("Mean Squared Error: {}".format(mean_absolute_error(Y_test, prediction)))
    print("Root Mean Squared Error: {}".format(np.sqrt(mean_squared_error(Y_test, prediction))))
```

# Modeling machine learning models

## Liner Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
prediction_linear_regretion=lr.predict(X_test)
```

```python
modelresults(prediction_linear_regretion)
```

```python
data["car purchase amount"].describe()
```

## Support Vector Machine

```python
from sklearn.svm import SVR
```

```python
svrmodel=SVR()
```

```python
param_gridsvr={"C":[0.01,0.1,0.5], "kernel": ["linear", "rbf", "poly"], "degree":[2,3,4]}
```

```python
from sklearn.model_selection import GridSearchCV
gridsvr=GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train, Y_train.values.ravel())
```

```python
prediction_gridsvr=gridsvr.predict(X_test)
```

```python
modelresults(prediction_gridsvr)
```

```python
gridsvr.best_params_
```

## Randome Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel=RandomForestRegressor()
param_gridrfr={"max_depth": [19,30,35,25,15,16,], "n_estimators":[30,40,50,60,70,80,90,100]}
```

```python
gridrfr=GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train, Y_train.values.ravel())
```

```python
prediction_rfr=gridrfr.predict(X_test)
```

```python
modelresults(prediction_rfr)
```
